### Random Forest Classifier

```python
from sklearn import ensemble
from sklearn.model_selection import cross_val_score
```

```python
rfc = ensemble.RandomForestClassifier()

X = y2015.drop('loan_status', 1)
Y = y2015['loan_status']
X = pd.get_dummies(X)

cross_val_score(rfc, X, Y, cv=5)
```
